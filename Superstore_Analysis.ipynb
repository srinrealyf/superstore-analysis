# Business Questions & Hypotheses

## Ship Mode vs Sum of Quantity

```python
ss_shipmode=ss_dataset[['Ship Mode', 'Category', 'Quantity']].groupby(['Ship Mode', 'Category']).sum('Quantity')
ss_shipmode
```

```python
ss_shipmode=ss_shipmode.sort_values(by='Quantity', ascending=False)
```

```python
ss_shipmode.plot(kind='bar', xlabel='Ship Mode - Category', ylabel='Quantity', title='Ship Mode vs Sum of Quantity', color='yellow')
```

## Top 10 Cities by Sales

```python
ss_dataset['City'].unique()
```

```python
city_sales=pd.DataFrame(ss_dataset.groupby('City')['Sales'].sum())
city_sales.reset_index(inplace=True)
```

```python
city_sales=city_sales.sort_values(by='Sales', ascending=False)
city_sales.head(10)
```

```python
city_sales_df=city_sales.head(10)
```

```python
city_sales_df.plot(kind='pie', y='Sales', labels=city_sales_df['City'], title='Top 10 Cities by Sales' ,autopct='%1.1f%%', legend=False)
plt.axis('off')
```

## Top 10 Cities by Profit

```python
ss_dataset['City'].unique()
```

```python
city_profit=pd.DataFrame(ss_dataset.groupby('City')['Profit'].sum())
city_profit.reset_index(inplace=True)
```

```python
city_profit=city_profit.sort_values(by='Profit', ascending=False)
city_profit.head(10)
```

```python
city_profit_df=city_profit.head(10)
city_profit_df
```

```python
city_profit_df.plot(kind='pie', y='Profit', labels=city_profit_df['City'], autopct='%1.1f%%', title='Top 10 Cities by Profit', legend=False)
plt.axis('off')
```

## Top 10 Cities by Sales vs Profit

```python
profit_sales = pd.merge(city_sales, city_profit, on="City", how="inner")
profit_sales.head(10)
```

```python
profit_sales_df=profit_sales.head(10)
profit_sales_df
```

```python
profit_sales_df.plot(kind='bar', x='City',title='Top 10 Cities by Sales vs Profit')
```

## Which is the most popular Sub-Category?

```python
popular_sub_category=pd.DataFrame(ss_dataset.groupby('Sub-Category')['Quantity'].sum())
popular_sub_category.sort_values(by='Quantity',ascending=False)
```

```python
popular_sub_category=popular_sub_category.sort_values(by='Quantity',ascending=False)
```

```python
popular_sub_category.plot(kind='bar', color='green', ylabel='Quantity',title='Most Popular Sub-Category', legend=False)
```

## Which State Have More Sales in Quantity?

```python
sales_quantity=pd.DataFrame(ss_dataset.groupby('State')['Quantity'].sum())
sales_quantity=sales_quantity.head(5).sort_values(by='Quantity',ascending=False)
```

```python
sales_quantity
```

```python
sales_quantity.plot(kind='line',y='Quantity',xlabel='State',ylabel='Quantity',title='States with Higher Sales in Quantity', figsize=(7,5), color='orange')
```

## Which States have Low Profits?

```python
low_profit=pd.DataFrame(ss_dataset.groupby('State')['Profit'].sum())
low_profit.head()
```

```python
low_profit=low_profit.head(5).sort_values(by='Profit', ascending=True)
```

```python
low_profit
```

```python
low_profit.plot(kind='line', xlabel='State', ylabel='Profits', title='States with Lower Profits', color='red')
```
